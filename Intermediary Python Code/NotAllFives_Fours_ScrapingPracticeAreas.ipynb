## This file scraped the practice areas for not all fours/fives

```python
import pandas as pd
import re

pd.set_option('display.max_colwidth', 150)
```

```python
df = pd.read_excel("NALP_2018_2021.xlsx", sheet_name = 'Not All Fives')
df
```

```python
#2018 To 2021 data
new_rows = []
matches = []
for idx,row in df.iterrows():
    general = row['FIND'] #GENERAL PRACTICE INFORMATION

    new_rows = re.findall("([a-zA-Z\s\_\-\&\,\.\/\+\'\n]+) [0-9]+", str(general)) #added period and /
    matches.append(new_rows)
        

df['test'] = matches
```

```python
df = df.explode('test').reset_index(drop=False) #drop = false right now to have unique firm ids
df
```

```python
df['Area'] = df.test.apply(lambda x: str(x).strip())

df[['index','Firm','test','Area']]
```
