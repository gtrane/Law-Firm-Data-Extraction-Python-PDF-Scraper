## This file scrape the numbers from the not all fours/fives.

```python
import pandas as pd
import re

pd.set_option('display.max_colwidth', 150)
```

```python
df = pd.read_excel("41423.xlsx")
df
```

```python
#2018 To 2021 data
new_rows = []
matches = []

for idx,row in df.iterrows():
    general = row['FIND'] #GENERAL PRACTICE INFORMATION

    new_rows = re.findall("([a-zA-Z\s\_\-\&\,\.\/\+\'\n]+\s\d{1,3}(\s\d{1,3}){0,4})", str(general))
    matches.append(new_rows)

df['test'] = matches
df
```

```python
df = df.explode('test').reset_index(drop=False) #drop = false right now to have unique firm ids
df
```

```python
df[['test']]
```

```python
#split test column into seven new columns

df['practice'] = df.test.apply(lambda x: re.findall("[a-zA-Z\s\_\-\&\,\.\/\+\'\n]+",str(x)))

df['numbers'] = df.test.apply(lambda x: re.findall("[0-9]+", str(x)))
df[['index','Firm','test','practice','numbers']]
```

```python
#couldn't get a pretty practice area scraping. just matched this column to my original file that didnt scrape numbers
# for "Not all fives/fours"
df['match'] = df.practice.apply(lambda x: " ".join(x).strip())
df[['match']]
```

```python
#new dataframe

# create a new dataframe with the values in separate columns
new_df = pd.DataFrame(df['numbers'].to_list(), columns=['1','2','3','4','5','6'])

# combine the new dataframe with the original dataframe
result = pd.concat([df, new_df], axis=1)
#result
```
